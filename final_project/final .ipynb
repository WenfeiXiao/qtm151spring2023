```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drivers = results = pd.read_csv("data_raw/drivers.csv")
results = pd.read_csv("data_raw/results.csv")
races = pd.read_csv("data_raw/races.csv")
```

```python
driver_merge = pd.merge(results,
                       drivers,
                       on = "driverId",
                       how = "left")

driver_merge 
driver_merge = driver_merge.drop(columns=['grid', 'url', 'code', 'number_y', 'number_x', 'positionText', 'positionOrder', 'laps'])
display (driver_merge)
```

```python
driver_race_merge = pd.merge(driver_merge,
                       races,
                       on = "raceId",
                       how = "left")

driver_race_merge = driver_race_merge.drop(columns=['url','position', 'fp1_date', 'fp1_time', 'fp2_date', 'fp2_time', 'fp3_date', 'fp3_time', 'quali_date','quali_time', 'round', 'circuitId', 'sprint_date', 'sprint_time','time_x'])

display (driver_race_merge)
```

```python
driver_race_merge["dob"] = pd.to_datetime(driver_race_merge["dob"])
driver_race_merge["age"] = driver_race_merge["year"].astype(int) - driver_race_merge["dob"].dt.strftime("%Y").astype(int)
```

```python
display (driver_race_merge)
```

```python
mean_point = driver_race_merge.groupby("age", as_index=False).agg(mean_point=("points", "mean")).sort_values(by="age")
```

```python
plt.scatter(x= mean_point["age"], y = mean_point["mean_point"])
plt.ylabel("Mean Points")
plt.xlabel("Age")
plt.title("Driver's success in relationship with age")
plt.show()
```

```python


plt.scatter(x= driver_race_merge["age"], y = driver_race_merge["points"])
plt.ylabel("Points")
plt.xlabel("Age")
plt.title("Driver's success in relationship with age")
plt.show()
```
